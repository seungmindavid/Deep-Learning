Remember $ϕ_i(x) = σ(w^0_i \cdot x)$
- When σ is a sigmoid function, $\sigma_i(x) = \frac{1}{1+e^{-w^0_i \cdot x}}$, which is a function that depends on W, and W will be updated through optimizer.

- A higher weight (w) value results in an increased steepness of the sigmoid function around the threshold.

Below plots are examples for showing how sigmoid reacts to different value of W


```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(w, x):
    return 1 / (1 + np.exp(-w * x))

x_values = np.linspace(-10, 10, 100)

# Plot the sigmoid function in the first subplot
y_values = sigmoid(1, x_values)

# Plot both the sigmoid function and its derivative in the same plot
plt.plot(x_values, sigmoid(1, x_values), label='Original Sigmoid')
plt.plot(x_values, sigmoid(10, x_values), label='w/ large positive')
plt.plot(x_values, sigmoid(0.5, x_values), label='w/ small positive')
plt.plot(x_values, sigmoid(-10, x_values), label='w/ large negative')
plt.plot(x_values, sigmoid(-0.5, x_values), label='w/ small negative')
plt.plot(x_values, sigmoid(0, x_values), label='w/ 0')
plt.title('Sigmoid Function with different W')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
